```python
#import
from tkinter import *
from tkinter import messagebox
import random

#base
root = Tk()
root.geometry('600x600')
root.title("tile matchimg game")
root.config(bg = "purple")
#num_list
num_list = ["a","a","s","s","f","f","d","d","y","y","p","p"]
random.shuffle(num_list)
#frame 
my_frame = Frame(root,bg ="purple")
my_frame.pack(pady = 45)

#def variables 
count = 0
answer_list = []
answer_dict = {}
winner = 0
record_score = 0

def score():
    global record_score
    lavel.config(text = record_score)
    



#Vdef win function
def win():
    label.config(text = "congratulations!! you won!!")
    splash = Tk()
    splash.title("congrats")
    splash.geometry(("300x300"))
    splash.config(bg = "blue")
    splash_label = Label(splash,text = "congrats on winning your game!!!!!", font = 20)
    splash_label.pack(pady = 20)
    button_list = [b0,b1,b2,b3,b4,b5,b6,b7,b8,b9,b10,b11]
    for button in button_list:
        button.config(bg = "#8fffa0")
    
#def buttons
def btn_click(b,number):
    global count
    global answer_list
    global answer_dict
    global winner
    global record_score
    
    if b["text"]== "" and count < 2:
        b["text"]= num_list[number]
        answer_list.append(number)
        answer_dict[b] = num_list[number]
        count += 1
        
        if len(answer_list)== 2:
            if num_list[answer_list[0]]==num_list[answer_list[1]]:
                
                record_score +=1
                if record_score> 0:
                    score()
                label.config(text = "match!")
                for key in answer_dict:
                    key["state"] = "disabled"
                    count = 0 
                    answer_list =[]
                    answer_dict = {}
                    winner+=1
                    if winner == 12:
                        win()
            else:
                #label.config(text = "uh-oh")
                count = 0 
                answer_list = []
                messagebox.showinfo("incorrect","incorrect")
                for key in answer_dict:
                    key["text"]= ""
                    answer_dict = {}
            

#buttons

lavel =Label(my_frame, text = "",font = 10,bg = "purple")
b0 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b0,0),bg = "#bb99f5",fg = "#392e5c")
b1 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b1,1),bg = "#bb99f5",fg = "#392e5c")
b2 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b2,2),bg = "#bb99f5",fg = "#392e5c")
b3 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b3,3),bg = "#bb99f5",fg = "#392e5c")

b4 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b4,4),bg = "#bb99f5",fg = "#392e5c")
b5 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b5,5),bg = "#bb99f5",fg = "#392e5c")
b6 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b6,6),bg = "#bb99f5",fg = "#392e5c")
b7 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b7,7),bg = "#bb99f5",fg = "#392e5c")

b8 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b8,8),bg = "#bb99f5",fg = "#392e5c")
b9 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b9,9),bg = "#bb99f5",fg = "#392e5c")
b10 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b10,10),bg = "#bb99f5",fg = "#392e5c")
b11 = Button(my_frame, text = "",font = ("helvetica",20),height = 3, width = 6,command = lambda:btn_click(b11,11),bg = "#bb99f5",fg = "#392e5c")
#placement of buttons with grid
lavel.grid(row = 1, column = 0)
b0.grid(row = 2, column = 0)
b1.grid(row = 2, column = 1)
b2.grid(row = 2, column = 2)
b3.grid(row = 2, column = 3)

b4.grid(row = 3, column = 0)
b5.grid(row = 3, column = 1)
b6.grid(row = 3, column = 2)
b7.grid(row = 3, column = 3)

b8.grid(row = 4, column = 0)
b9.grid(row = 4, column = 1)
b10.grid(row = 4, column = 2)
b11.grid(row = 4, column = 3)

button1 = Button(my_frame, text = "")

label = Label(my_frame,text = "",font=20,bg = "purple")
label.grid( rows = 5 , column = 2,)

label1 = Label(my_frame,text = "briefing: click on buttons toget ",bg = "purple")
label1.grid(rows = 6,column = 0)

label2 = Label(my_frame,text = "letters and try to match them",bg = "purple")
label2.grid(rows = 7,column = 0)
#menu function
def reset():
    global num_list
    global winner
    global record_score
    record_score  = 0
    lavel.config(text = record_score)
    winner = 0
    num_list = ["a","a","s","s","f","f","d","d","y","y","p","p"]
    random.shuffle(num_list)
    label.config(text = "")
    button_list = [b0,b1,b2,b3,b4,b5,b6,b7,b8,b9,b10,b11]
    for button in button_list:
        button.config(text = "",bg = "#bb99f5",state = "normal")
    
    
    
    
#menu 
my_menu = Menu(root)
root.config(menu = my_menu)
option_menu = Menu(my_menu,tearoff = False)
my_menu.add_cascade(label = "options", menu = option_menu)
option_menu.add_command(label = "reset game",command = reset)


root.mainloop()
```
